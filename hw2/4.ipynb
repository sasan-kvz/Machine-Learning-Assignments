```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as skl
```

```python
x = np.arange(-10 , 10 , 0.2)
y = np.zeros(100)
for i in range(100):
    y[i] = 2 * (math.cos(x[i]))/-math.pi + 2 * math.sin(2 * x[i])/(2 * math.pi) + 2 * math.cos(3 * x[i])/(-3 * math.pi)

white_noise = np.random.normal(0, 1, size=100)    
y1 = y+ 0.12 *white_noise
poisson_noise =  np.random.poisson(2,100)
y2 = y+ 0.12 *poisson_noise

plt.scatter(x,y1)
plt.xlabel("x", fontsize=15)
plt.ylabel("y1",rotation=0 ,fontsize=15)
plt.title("white noise")
plt.figure()
plt.scatter(x,y2)
plt.xlabel("x", fontsize=15)
plt.ylabel("y2",rotation=0 ,fontsize=15)
plt.title("poisson noise")
```

```python
degree= np.arange(1,16,1)
z = {d:0 for d in degree}
fit = {d:0 for d in degree}
y_pred = {d:0 for d in degree}
MSE = {d:0 for d in degree}
variance = {d:0 for d in degree}
bias= {d:0 for d in degree}
for i in degree:
    z[i]= np.polyfit(x,y1,i)
    fit[i] = np.poly1d(z[i])
    plt.figure()
    plt.scatter(x,y1)
    plt.plot(x, fit[i](x))
    plt.xlabel("x", fontsize=15)
    plt.ylabel("y1_predicted",fontsize=15)
    plt.title("degree of polynomial = {}".format(i))
    plt.grid()
    y_pred[i] = fit[i](x)
    MSE[i] = skl.metrics.mean_squared_error(y,y_pred[i])
    variance[i] = np.var(y_pred[i])
    bias[i] = np.sqrt(np.abs(MSE[i] - variance[i]))
print(MSE)
print(variance)
print(bias)
```

```python
plt.plot(degree,list(MSE.values()), label="MSE1")
plt.plot(degree,list(variance.values()), label="Variance1")
plt.plot(degree,list(bias.values()), label="Bias2")
plt.xlabel("degree of polynomial")
plt.ylabel("parameters")
plt.legend(fontsize=8)
plt.show()
```
